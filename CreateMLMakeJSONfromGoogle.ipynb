```python
import os
import json
from PIL import Image
```

```python
def chunks(l, n):
    n = max(1, n)
    return list(l[i:i+n] for i in range(0, len(l), n))
```

```python
clist=[]
data_one = {}
jsonfile=[]

os.chdir(os.path.join("/Users/pano/Downloads/OIDv4_Dress/OID/Dataset/train/label/"))

for i in os.listdir(os.getcwd()): 
    f = open(i, "r")
    name =i.split('.')[0]


    coordinates=f.read().split()
    coordinates=list(set(coordinates))
    try:
        coordinates.remove('Dress')
        coordinates=[float(x) for x in coordinates]
    except:
        pass
    
    coord_chunks=chunks(coordinates,4)
    data_one['imagefilename'] = name +".jpg"
    an_list=[]
    for i in coord_chunks:   
        if len(i)==4:
            
            counter=0
            x_center = i[0]
            y_center = i[1]
            width = i[2]
            height = i[3]

            cdict={}
            cdict['y']=y_center
            cdict['x']=x_center
            cdict['height']=height
            cdict['width']=width
            
            an_list.append({'label':"Dress",'coordinates':cdict})
            counter+=1
            
    data_one['annotations']=an_list
    jsonfile.append(data_one.copy())
```

```python
with open('/Users/pano/python/annotations.json', 'w') as outfile:
    json.dump(jsonfile, outfile)
```
